```python
#checking missing values and its sum in each columns
df.isnull().sum()
```

```python
#Checking how much percent of missing values are in the columns
percent_missing_in_establishment= (df["establishment"].isnull().sum()/df["establishment"].count())*100
print(f" the missing percent in establishment is :{percent_missing_in_establishment:.2f} percent")
print("\n")
percent_missing_in_address= (df["address"].isnull().sum()/df["address"].count())*100
print(f" the missing percent in address is :{percent_missing_in_address:.2f} percent")
print("\n")
percent_missing_in_zipcode= (df["zipcode"].isnull().sum()/df["zipcode"].count())*100
print(f" the missing percent in zipcode is :{percent_missing_in_zipcode:.2f} percent")
print("\n")
percent_missing_in_cuisines= (df["cuisines"].isnull().sum()/df["cuisines"].count())*100
print(f" the missing percent in cuisines is :{percent_missing_in_cuisines:.2f} percent")
print("\n")
percent_missing_in_timings= (df["timings"].isnull().sum()/df["timings"].count())*100
print(f" the missing percent in timings is :{percent_missing_in_timings:.2f} percent")
print("\n")
percent_missing_in_highlights= (df["highlights"].isnull().sum()/df["highlights"].count())*100
print(f" the missing percent in highlights is :{percent_missing_in_highlights:.2f} percent")
print("\n")
percent_missing_in_opentable_support= (df["opentable_support"].isnull().sum()/df["opentable_support"].count())*100
print(f" the missing percent in opentable_support is :{percent_missing_in_opentable_support:.2f} percent")
print("\n")
```

### with reference to above cell o/p 
for most of the missing percent, percentage is less than 5% expect for zipcode which has missing percent of ~334%

```python
# for less than 5% of missing values, we are deleting those rows as deleting it dont have much impact on our data in term of size
#for zipcode we will be using impute function and fillna by MODE
df['zipcode'].fillna(df['zipcode'].mode()[0], inplace=True)
df['zipcode'].isna().sum()
```

```python
#dropping all nulls from all columns
df.dropna(inplace=True)
df.info()
```

```python
#grouping dataframe on price_range
df_groupby_price_range=df.groupby('price_range')['average_cost_for_two'].mean()
df_groupby_price_range
```

```python
#ploting bar chart horizontal to show average cost by price range
plt.figure(figsize=(6,8))
plt.barh(df['price_range'],df['average_cost_for_two'],color='red')
plt.title("Average cost by price range")
plt.ylabel("Price range class")
plt.xlabel("Average Price for two people")
plt.grid()
```

# inference from above chart
for upto class 3 price gradually increases but for premimum class (4) they charge very high compared to other three class

```python
#ploting bar chat to show average cost of food by city
plt.figure(figsize=(6,25))
plt.barh(df['city'],df['average_cost_for_two'],color='blue')
plt.title("Average cost by city")
plt.xlabel("Average cost for food")
plt.ylabel("City")
plt.grid()
plt.show()
```

### with reference to above cell o/p we can see top 5 (topN) city by most expensive city by average cost of food for two

```python
#filtering whole dataframe by rating more than 4
df_rating_more_than4= df[df['aggregate_rating']>4]
df_rating_more_than4.duplicated().any()
```

```python
#average cost of food by city for rating more than 4
plt.figure(figsize=(6,20))
plt.barh(df_rating_more_than4['city'],df_rating_more_than4['average_cost_for_two'], color='skyblue')
plt.title("Average cost of food in city which has more than 4 rating")
plt.xlabel("City")
plt.ylabel("Average cost of food for two person")
plt.show()
```

```python
#outliers using scatter plot
plt.figure(figsize=(10,10))
sns.scatterplot(x=df_rating_more_than4['photo_count'],y=df_rating_more_than4['votes'], hue='average_cost_for_two', data=df_rating_more_than4)
plt.ylabel("Votes")
plt.xlabel("Number of photo posted")
plt.show()
```

```python
#matrix visual showing mean of avg cost by city and for rating more than 4
matrix = pd.pivot_table(
    df_rating_more_than4,
    index='city',          
    columns='aggregate_rating',
    values='average_cost_for_two',
    aggfunc='mean'
).fillna(0)

matrix.style.background_gradient(cmap='Blues').format("{:.2f}")
```

```python
#matrix visual showing mean of avg cost by establishment and rating
matrix2 = pd.pivot_table(
    df_rating_more_than4,
    index='establishment',
    columns="aggregate_rating",
    values="average_cost_for_two",
    aggfunc='mean'
).fillna(0)
matrix2.style.background_gradient(cmap="BuGn").format("{:.2f}")
```

```python
#photo count and vote relation ship
plt.figure(figsize=(10,10))
sns.scatterplot(x=df_rating_more_than4['photo_count'], y=df_rating_more_than4['votes'],hue='aggregate_rating',data=df_rating_more_than4)
#higher votes and photo clicked, higher the rating
```

```python
#correlation data frame
df_for_correlation= df[['average_cost_for_two', 'aggregate_rating' ,'votes','photo_count']] 
df_for_correlation
```

```python
#measuring co relation between below four cols names
correlation_= df_for_correlation.corr(method='pearson')
correlation_
```

```python
# Heat map -- for showing correlation between above four columns

plt.figure(figsize = (10, 10)) 
sns.heatmap(correlation_.corr(), annot = True, cmap="viridis") 
```
